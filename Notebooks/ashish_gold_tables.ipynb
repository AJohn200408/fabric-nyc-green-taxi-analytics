### Silver to Gold

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
silver_df = spark.table("ashish_silver_trips")
silver_df.columns
```

```python
dim_pickup = silver_df.select(
    col("pickup_location_id")
).distinct()

dim_pickup = dim_pickup.withColumn(
    "pickup_location_name",
    concat(
        lit("Location_"),
        col("pickup_location_id")
    )
)
```

```python
dim_pickup.write.mode("overwrite").saveAsTable(
    "ashish_gold_dim_pickup_location"
)
```

```python
dim_dropoff = silver_df.select(
    col("dropoff_location_id")
).distinct()

dim_dropoff = dim_dropoff.withColumn(
    "dropoff_location_name",
    concat(
        lit("Location_"),
        col("dropoff_location_id")
    )
)
```

```python
dim_dropoff.write.mode("overwrite").saveAsTable(
    "ashish_gold_dim_dropoff_location"
)
```

```python
fact_trips = silver_df.select(

    col("vendor_id"),

    col("pickup_location_id"),


    date_format(
        col("pickup_datetime"),
        "yyyyMMdd"
    ).cast("int").alias("pu_date_key"),


    col("dropoff_location_id"),


    date_format(
        col("dropoff_datetime"),
        "yyyyMMdd"
    ).cast("int").alias("do_date_key"),

    concat(
        col("pickup_location_id"),
        lit(" → "),
        col("dropoff_location_id")
    ).alias("pickup_dropoff_pair"),

    col("passenger_count"),

    col("trip_distance"),

    col("fare_amount"),

    col("tip_amount"),

    col("tolls_amount"),

    col("total_amount")
)
```

```python
fact_trips.write.mode("overwrite").saveAsTable(
    "ashish_gold_fact_trips"
)
```

```python
print(f"Rows in silver table = {silver_df.count()} \n Rows in fact table = {fact_trips.count()} ")
print(f"Rows in dim_pickup table = {dim_pickup.count()} \n Rows in dim_dropoff table = {dim_dropoff.count()} ")
```
