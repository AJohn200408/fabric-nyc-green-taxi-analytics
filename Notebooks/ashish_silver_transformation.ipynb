 ### Handling Null values

```python
drop_columns = [
    'ehailFee',
    'pickupLatitude',
    'pickupLongitude',
    'dropoffLatitude',
    'dropoffLongitude'
]

df = df.drop(columns=drop_columns)
```

```python
df = df.dropna(subset=['puLocationId', 'doLocationId'])
```

```python
print(df.columns)
```

### Validity rules
1. numeric values > 0 (passengerCount', 'tripDistance', 'fareAmount', 'extra',
       'mtaTax', 'improvementSurcharge', 'tipAmount', 'tollsAmount',
       'totalAmount')

2. Limit max(datetime) to current
3. puLocationId and doLoacationId is not null
4. Drop off Time > Pick up time 

```python
numeric_columns = [
    'passengerCount',
    'tripDistance',
    'fareAmount',
    'extra',
    'mtaTax',
    'improvementSurcharge',
    'tipAmount',
    'tollsAmount',
    'totalAmount'
]

for col in numeric_columns:
    df[col] = df[col].abs()
```

```python
df = df[
    (df["tripDistance"] > 0) &
    (df["totalAmount"] > 0) &
    (df["passengerCount"] > 0)
]

print("After invalid values removed:", df.shape[0])
```

### Limit to current date and time.

```python
current_date = pd.Timestamp.now()

df = df[df['lpepPickupDatetime'] <= current_date]

df = df[df['lpepDropoffDatetime'] <= current_date]
```

### Drop off time > Pick up time

```python
df = df[df['lpepDropoffDatetime'] > df['lpepPickupDatetime']]
```

### Creating Date Dimension Table.

Extract dates

```python
pickup_dates = silver_df['pickup_datetime'].dt.date

dropoff_dates = silver_df['dropoff_datetime'].dt.date
```

```python
all_dates = pd.concat([
    pickup_dates,
    dropoff_dates
]).drop_duplicates().reset_index(drop=True)
```

```python
dim_date = pd.DataFrame({
    'date': pd.to_datetime(all_dates)
})
```

```python
dim_date['date_key'] = dim_date['date'].dt.strftime('%Y%m%d').astype(int)

dim_date['year'] = dim_date['date'].dt.year

dim_date['quarter'] = dim_date['date'].dt.quarter

dim_date['month'] = dim_date['date'].dt.month

dim_date['month_name'] = dim_date['date'].dt.month_name()

dim_date['day'] = dim_date['date'].dt.day

dim_date['day_name'] = dim_date['date'].dt.day_name()

dim_date['weekday'] = dim_date['date'].dt.weekday
```

```python
#Re-ordering the columns
dim_date = dim_date[
    [
        'date_key',
        'date',
        'year',
        'quarter',
        'month',
        'month_name',
        'day',
        'day_name',
        'weekday'
    ]
]
```
